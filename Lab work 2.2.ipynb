```python
numbers = input("Введіть три числа через пробіл")
a, b, c = numbers.split()
a = int(a)
b = int(b)
c = int(c)
print("max")
if a>b:
    print(a) if a>c else print(c)
else: 
    print(b) if b>c else print(c)
print("min")
if a<b:
    print(a) if a<c else print(c)
else:
    print(b) if b<c else print(c)
    

```

Напишіть функції, що:
* рохраховує мінімальне та максимальне значення
* середнє значення послідовності
* медіану послідовності
* частоту кожної цифри в послідовності
* середнє квадратичне відхилення

Послідовність чисел ганеруємо рандомно

```python
import random
from statistics import stdev
from statistics import median

#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]

print("мінімальне заначення", min(list_with_random_numbers))       
print("максимальне заначення", max(list_with_random_numbers))      
print("середнє значення", sum(list_with_random_numbers)/len(list_with_random_numbers))
print("медіана послідовності", median(list_with_random_numbers))
print("середнє квадратичне відхилення", stdev(list_with_random_numbers))


```
